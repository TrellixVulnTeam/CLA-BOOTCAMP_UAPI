```python
! pip install mlxtend 
import import_ipynb
import random
import string
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
from mlxtend.preprocessing import TransactionEncoder
random.seed(10)


```

```python
vocab=['A', 'B', 'C', 'D', 'E']
max_length=len(vocab)
N=12
```

```python
def create_dataset(N, vocab, max_length):
    result=[]
    for i in range (N):
        n_items= random.randint(0,max_length)
        result.append([random.choice(vocab) for i in range (n_items)])
    return result 
    
        
        
```

```python
data = create_dataset(N, vocab, max_length)
print(data)
```

```python
# Let's transform the list with one hot-encoding 
a= TransactionEncoder()
a_data = a.fit(data).transform(data)
df1= pd.DataFrame(a_data, columns= a.columns_)
df1 = df1.replace(False, 0)
df1
```

```python
df= apriori(df1, min_support = 0.3, use_colnames= True, verbose= 1)
df
```
